```python
from fr.spikes import TuningCurve, NonConstantSpikeTrain, SpikeTrain
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

tuncurve = TuningCurve(r_max=5, s_max=20, σ=20, range=[-40, 40])
spikes = NonConstantSpikeTrain(T = 100, Δt = 0.1, 
                               tuncurve=tuncurve)
spikes.spike_train = spikes.gen_spike_train()
spikes.plot_spike_train()
```

```python
spikes.t
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
aprox_rt = spikes.gaussian_delta_filter(2)
sns.lineplot(x=spikes.t, y=aprox_rt)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
aprox_rt = spikes.linear_delta_filter(30)
sns.lineplot(x=spikes.t, y=aprox_rt)
```
